# (2) Robustness results under different identification threshold

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# data
data = [97.91, 97.52, 97.37, 97.69]


plot_df = pd.DataFrame({
    'Threshold': ['$2 \\times Std_{RIC}$', '$3 \\times Std_{RIC}$', '$4 \\times Std_{RIC}$', '$5 \\times Std_{RIC}$'],
    'Percentage': data
})


fig, ax = plt.subplots(figsize=(6, 4))


ax.plot(data, marker='o', color='r', lw=3)  


ax.yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(xmax=100, decimals=0))


ax.set_xlabel('Peer identification threshold $\delta$', fontsize=15)
ax.set_ylabel('Percentage', fontsize=15)

ax.set_ylim(60, 100)
ax.set_yticks(np.arange(60, 101, 10))

x = np.arange(len(data))
ax.set_xticks(x)
ax.set_xticklabels(['$2 \\times Std_{IC}$', '$3 \\times Std_{IC}$', '$4 \\times Std_{IC}$', '$5 \\times Std_{IC}$'], fontsize=15, fontweight='bold')

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(15)
    label.set_fontweight('bold')


plt.tight_layout()
plt.savefig('unfairly_treated_threshold_robustness.pdf', format='pdf', dpi=300)
plt.show()
plot_df.to_csv('unfairly_treated_threshold_robustness.csv', index=False)
```
